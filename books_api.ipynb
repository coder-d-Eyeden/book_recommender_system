```python
import requests

def search_openlibrary(book_name):

    url = f"https://openlibrary.org/search.json?title={book_name}"

    response = requests.get(url)

    data = response.json()

    if data["numFound"] == 0:
        return None

    book = data["docs"][0]

    return {
        "title": book.get("title"),
        "author": book.get("author_name", ["Unknown"])[0],
        "year": book.get("first_publish_year")
    }
```

```python
print(search_openlibrary("1984"))
```
